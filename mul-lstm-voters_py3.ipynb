```python
import glob
import gc
import numpy as np
from matplotlib import pyplot
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import LSTM
```

```python
import time
start = time.time()
path = 'C:/Users/inbar&aryeh/Documents/academy/anomaly detection/cic/CSVs/' # use your path
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
models = []
for file_ in allFiles:
    print(file_)
    
    try:
        frame = pd.read_csv(file_,index_col=None, header=0)
    except:
        frame = pd.read_csv(file_,index_col=None, header=0, encoding = 'latin1')
    frame = frame.dropna(how = 'all')
    frame[' Flow Packets/s'] = frame[' Flow Packets/s'].astype('float64')
    frame['Flow Bytes/s'] = frame['Flow Bytes/s'].astype('float64')
    frame = frame.drop(' Timestamp', axis = 1)
    frame = frame.replace([np.inf, -np.inf], np.nan)
    non_trainable = ['Flow ID', ' Source IP', ' Destination IP', ' Timestamp', 'External IP', ' Label', 'dest_a','dest_b','dest_c','dest_d','src_a','src_b', 'src_c','src_d']
    label_names = list(frame[' Label'].drop_duplicates())
    label_mapping = {val: i for i, val in enumerate(label_names)}
    print(label_names)
    
    for name in label_names[1:]:
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~ " + name + " ~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        
        # split scale and reshape
        frame_tag = frame.loc[frame[' Label'].isin([label_names[0], name])]
        frame_mat = frame_tag[[col for col in frame_tag.columns if col != ' Label' and col not in non_trainable]].as_matrix()
        y_binary = frame_tag[' Label'].apply(lambda val: 1 if val != 'BENIGN' else 0).as_matrix()
        seq_length = 2
        dataX = []
        dataY = []
        dataX = [frame_mat[i:i + seq_length] for i in range(len(frame_mat) - seq_length)]
        dataY = [y_binary[i + seq_length - 1] for i in range(len(frame_mat) - seq_length)]
        data_X_tsteps = np.array(dataX)
        data_y_tsteps = np.array(dataY)
        X_train, X_test, y_train, y_test = train_test_split(np.nan_to_num(data_X_tsteps), data_y_tsteps, test_size=0.2, shuffle=True, random_state=42, stratify=data_y_tsteps)
        
        #scaler = StandardScaler()
        #scaled_train_X  = scaler.fit_transform(X_train[[column for column in X_train.columns if column not in non_trainable]].fillna(0))
        #scaled_test_X = scaler.transform(X_test[[column for column in X_test.columns if column not in non_trainable]].fillna(0))
        #scaled_train_X  = scaled_train_X.reshape((scaled_train_X.shape[0], 1, scaled_train_X.shape[1]))
        #scaled_test_X = scaled_test_X.reshape((scaled_test_X.shape[0], 1, scaled_test_X.shape[1]))
        
        # design network
        model = Sequential()
        model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])))
        model.add(Dense(32, activation = 'relu'))
        model.add(Dense(1,  activation='sigmoid'))
        model.compile(loss='binary_crossentropy', optimizer='adam')
        print(model.summary())
        # fit network
        es = EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience=3,
                              verbose=0, mode='auto')
        history = model.fit(X_train, y_train, epochs=100, batch_size=72, validation_data=(X_test, y_test), verbose=2, shuffle=True, callbacks=[es])
        models.append(model)
        # plot history
        pyplot.plot(history.history['loss'], label='train')
        pyplot.plot(history.history['val_loss'], label='test')
        pyplot.legend()
        pyplot.show()
        preds = model.predict(X_test)
        print(classification_report(y_test, [1 if val > 0.5 else 0 for val in preds]))
    gc.collect()
end = time.time()
print(end - start)
```

```python
start = time.time()
frame = pd.DataFrame()
maxs = None
avgs = None
mins = None
y = pd.DataFrame()
j = 0
for file_ in allFiles:
    print(file_)
    
    try:
        frame = pd.read_csv(file_,index_col=None, header=0)
    except:
        frame = pd.read_csv(file_,index_col=None, header=0, encoding = 'latin1')
    frame = frame.dropna(how = 'all')
    frame[' Flow Packets/s'] = frame[' Flow Packets/s'].astype('float64')
    frame['Flow Bytes/s'] = frame['Flow Bytes/s'].astype('float64')
    frame = frame.drop(' Timestamp', axis = 1)
    frame = frame.replace([np.inf, -np.inf], np.nan)
    label_names = list(frame[' Label'].drop_duplicates())
    label_mapping = {val: i for i, val in enumerate(label_names)}
    print(label_names)
    
    for name in label_names[1:]:
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~ " + name + " ~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        frame_tag = frame.loc[frame[' Label'].isin([label_names[0], name])]
        frame_mat = frame_tag[[col for col in frame_tag.columns if col != ' Label' and col not in non_trainable]].as_matrix()
        #y_binary = frame_tag[' Label'].apply(lambda val: 1 if val != 'BENIGN' else 0).as_matrix()
        seq_length = 2
        dataX = []
        #dataY = []
        dataX = [frame_mat[i:i + seq_length] for i in range(len(frame_mat) - seq_length)]
        #dataY = [y_binary[i + seq_length - 1] for i in range(len(frame_mat) - seq_length)]
        data_X_tsteps = np.array(dataX)
        #data_y_tsteps = np.array(dataY)
        y_cur = frame_tag[' Label']
        y = pd.concat([y, y_cur.apply(lambda val: 1 if val == name else 0)])
        
        pred_mat = []
        for i, model in enumerate(models):
            if i != j:
                pred_mat.append(model.predict(data_X_tsteps))
        pred_mat = np.array(pred_mat)
        maxs = pred_mat.max(0) if maxs is None else np.concatenate([maxs, pred_mat.max(0)])
        avgs = pred_mat.sum(0) / len(pred_mat) if avgs is None else np.concatenate([avgs, pred_mat.sum(0) / len(pred_mat)])
        mins = pred_mat.min(0) if mins is None else np.concatenate([mins, pred_mat.min(0)])
        j += 1
print(time.time() - start)
```

```python
from sklearn.metrics import roc_curve, auc

def plt_roc(fpr, tpr):
    pyplot.figure()
    lw = 2
    pyplot.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % auc(fpr, tpr))
    pyplot.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    pyplot.xlim([0.0, 1.0])
    pyplot.ylim([0.0, 1.05])
    pyplot.xlabel('False Positive Rate')
    pyplot.ylabel('True Positive Rate')
    pyplot.title('AUC-ROC curve')
    pyplot.legend(loc="lower right")
    pyplot.show()
```

```python
preds = [1 if (m == 1) else 0 for m, a in zip(maxs, avgs)]
l = min(len(preds), len(y))
print(classification_report(y[:l], preds[:l]))
fpr, tpr, thresholds = roc_curve(y[:l], preds[:l])
plt_roc(fpr, tpr)
```
